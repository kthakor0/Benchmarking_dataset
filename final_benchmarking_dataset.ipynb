the perpose of this code is to clean up the output from Legacy Workflow and Bactopia, merge them togther and get the common and unique gene and plasmid IDs and at the end create a final DataFrame that has all the output from both of the tools as well the common and unique IDS from them

```python
import pandas as pd
import os 
import numpy as np
import re
```

read the datatframe as excel sheet

```python
table_1 = pd.read_excel("amrfinderplus-genes_1.xlsx")
table_1
```

select only the rows which has scop=core

```python
final_table = table_1.query('Scope == "core"')
final_table
```

Since only interested columns are  Name(sample_ID) and respective Gene symbol, selecting only these two column to work with 

```python
final_table = final_table[['Name','Gene symbol']]
final_table = final_table.astype(str)
final_table
```

Above datatfrane has repeated sample ID with each gene sy,mbol in each row. I am merging them all genesymbol with its respective sample Id seperated by coma 

```python
grouped = final_table.groupby('Name').agg({'Gene symbol':lambda x:','.join(x)})

grouped
```

replacing '_' with '(' and closing the bracket at the end of the ID. so that the requirement of recognized common ID can be fulfilled without any mistake 

```python
grouped.to_excel("Bactopia_out.xlsx")
```

```python
grouped['Gene symbol'] = grouped['Gene symbol'].apply(lambda x: np.array(x.split(','))if isinstance(x, str) else x)
grouped['Gene symbol'] = grouped['Gene symbol'].apply(lambda x: [elem.replace('_', '(') + ')' if isinstance(elem, str) and '_' in elem else elem for elem in x])
grouped['Gene symbol'] = grouped['Gene symbol'].apply(lambda x: ','.join(x) if isinstance(x, list) else '')
grouped
```

```python
 dir_path = "/scicomp/home-pure/tgu0/GWA/EDLB/projects/Krishna/Benchmarking/Benchmarking/BactopiaAMR/bactopia-tools/plasmidfinder/plasmidfinder/final_tsv"
```

merge the exixting and the new .tsv file (first converting it in to dataframe) together so we can get the common and noncommon gene and plasmid IDS 

```python
merged_df = pd.DataFrame(columns = ["Name"])
```

```python
for filename in os.listdir(dir_path):
    if filename.endswith(".tsv"):
        # Read in the TSV file as a DataFrame
        tsv_df = pd.read_csv(os.path.join(dir_path, filename), sep="\t", header=None)

        # Add a new column to the DataFrame with the filename
        tsv_df["Name"] = filename

        # Append the DataFrame to the merged DataFrame
        merged_df = merged_df.append(tsv_df, ignore_index=True)

# Write the merged DataFrame to a new TSV file without headers
merged_df.to_csv("merged_file.tsv", sep="\t", index=False, header=False)
merged_df
        
```

```python

merged_df.columns = ["Name","Database","Plasmid","Identity","Query / Template length","Contig","Position in contig","Note","Accession number"]

merged_df
```

reset index 

```python
for i in range(len(merged_df)):
    if merged_df.loc[i, "Database"] == "Database":
        merged_df = merged_df.drop([i], axis=0)
merged_df = merged_df.reset_index()
merged_df
```

drop .tsv sufix from name column elements and grouped them together so we have nice coma sepeated plasmid ID respective to their Name 

```python
merged_grouped = merged_df.groupby('Name').agg({'Plasmid':lambda x:','.join(x)}) #group the Name and Plasmid column together and seperate the plasmid ID by ','
merged_grouped = merged_grouped.reset_index()
merged_grouped = merged_grouped.drop(merged_grouped.index[0]) #drop .tsv from the name  
merged_grouped['Name'] = merged_grouped['Name'].str.replace('.tsv','')
print(merged_grouped)
```

read the excel file as .xlsx file nad na,e the column and drop the last unwanted column 

```python
workFlow_out= pd.read_excel("final.summary.salm.2023_01_30_17_04_03.xlsx") #read the excel file 
workFlow_out.columns = ["Name","gene","plasmid_workFlow","NARMS"]#name columns
del workFlow_out['NARMS']#drop the NARMS column
workFlow_out
```

merged the two output dataframe all together to work with so now we cna get the common elements from both output which is from workflow and bactopia 

```python
final_df = pd.merge(pd.merge(workFlow_out,merged_grouped, on='Name', how='outer'),grouped,on='Name', how='outer') #merge the DataFrame based on common column(Name)
final_df = final_df.fillna('NaN') #replacing empty spaces with NaN
final_df = final_df.replace('None','NaN') #replacing None with NAN
final_df
```

```python
final_df.to_excel("banchmarkingFinal.xlsx") #export the dataframe as excel sheet
```

```python
final_df['gene_elements'] = final_df['gene'].str.split(',') #excess each of the gene colum element 
final_df['Gene_symbol_elements']= final_df['Gene symbol'].str.split(',') # excess each of the comamasepearted element of the gene_symbol_element 
final_df['gene_set'] = final_df['gene_elements'].apply(set)# excess each elemnts of the gene_elements
final_df['gene_symbol_set'] = final_df['Gene_symbol_elements'].apply(set)#excess gene symbol elemts 
```

get the common and unique gene_IDS from workflow and bactopia output

```python
final_df['common_elemets'] = final_df.apply(lambda row: row['gene_set'].intersection(row['gene_symbol_set']),axis=1) # get the common gene_ID  from workflow and Bactopia 
final_df['missingFrom_gene'] = final_df.apply(lambda row: row['gene_symbol_set']-row['gene_set'],axis=1)# get he missing gene_IDS from 'gene' column
final_df['missingFrom_geneSymbol'] = final_df.apply(lambda row: row['gene_set']-row['gene_symbol_set'],axis=1)# get the miising gene IDS from 'genesymbol' column
final_df
```

```python
final_df['plasmidBactopia_elements'] = final_df['Plasmid'].str.split(',')
final_df['plasmidWorkflow_elements']= final_df['plasmid_workFlow'].str.split(',')
final_df['plasmidBactopia_set'] = final_df['plasmidBactopia_elements'].apply(set)
final_df['plasmidWorkflow_set'] = final_df['plasmidWorkflow_elements'].apply(set)
```

```python
final_df['commonPlasmidelemets'] = final_df.apply(lambda row: row['plasmidBactopia_set'].intersection(row['plasmidWorkflow_set']),axis=1)
final_df['missingFrom_plasmidBactopia'] = final_df.apply(lambda row: row['plasmidWorkflow_set']-row['plasmidBactopia_set'],axis=1)
final_df['missingFrom_PlasmidWorkflow'] = final_df.apply(lambda row: row['plasmidBactopia_set']-row['plasmidWorkflow_set'],axis=1)
final_df
```

remane the final datatframe which can make sense 

```python
banchmarkingFinal = final_df[['Name','gene','Gene symbol','common_elemets','missingFrom_gene','missingFrom_geneSymbol','Plasmid','plasmid_workFlow','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']]
banchmarkingFinal
```

replacing '{}' with ' ' [empty space]:(get rid of all {} from dataframe)

```python
banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']] = banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']].apply(lambda x: x.astype(str).str.replace('[{}]',' '))
banchmarkingFinal
```

```python
banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']] = banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']].apply(lambda x: x.astype(str).str.replace('[{}]',''))
banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']] = banchmarkingFinal[['common_elemets','missingFrom_gene','missingFrom_geneSymbol','commonPlasmidelemets','missingFrom_plasmidBactopia','missingFrom_PlasmidWorkflow']].apply(lambda x: x.astype(str).str.replace('[""]',''))
banchmarkingFinal
```

replace "'" with empty space 

```python
banchmarkingFinal = banchmarkingFinal.applymap(lambda x: re.sub(r"^\s*'|\s*'$", '', x))
banchmarkingFinal['common_elemets'] = banchmarkingFinal['common_elemets'].astype(str).apply(lambda x: x.replace("'","")) # replace "'" with empty space 
banchmarkingFinal
```

remove "'" from all respective colum values end

```python
banchmarkingFinal['missingFrom_gene'] = banchmarkingFinal['missingFrom_gene'].apply(lambda x: re.sub('[^0-9a-zA-Z+]+', ',', x))
banchmarkingFinal['missingFrom_plasmidBactopia'] = banchmarkingFinal['missingFrom_plasmidBactopia'].apply(lambda x: re.sub('[^0-9a-zA-Z+]+', ',', x))
banchmarkingFinal['commonPlasmidelemets'] = banchmarkingFinal['commonPlasmidelemets'].apply(lambda x: re.sub('[^0-9a-zA-Z+]+', ',', x))
banchmarkingFinal['missingFrom_PlasmidWorkflow'] = banchmarkingFinal['missingFrom_PlasmidWorkflow'].apply(lambda x: re.sub('[^0-9a-zA-Z+]+', ',', x))
banchmarkingFinal['missingFrom_geneSymbol'] = banchmarkingFinal['missingFrom_geneSymbol'].apply(lambda x: re.sub('[^0-9a-zA-Z+]+', ',', x))
banchmarkingFinal
```

```python
banchmarkingFinal = banchmarkingFinal.replace({'set,', 'set()', ' '}, np.nan) # replace set and set() with empty space 
banchmarkingFinal.iloc[:, -2:] = banchmarkingFinal.iloc[:, -2:].replace(r'^\s*$', np.nan, regex=True) # to replace all empty space with NaN
banchmarkingFinal
```

rename all the colum of the final DataFrame so that we can understan d what represent what 

```python
banchmarkingFinal = banchmarkingFinal.rename(columns={'Name':'Sample ID','gene':'Gene_ID Workflow', 'Gene symbol':'Gene_Id Bactopia', 'common_elemets':'Common Genes','missingFrom_gene':'Missing From Workflow (Gene)','missingFrom_geneSymbol':'Missing From Bactopia','Plasmid':'Plasmid_ID Bactopia','plasmid_workFlow':'Plasmid_ID  workFlow','commonPlasmidelemets':'common Plasmid IDs','missingFrom_plasmidBactopia':'Missing_Ids From Bactopia (Plasmid)','missingFrom_PlasmidWorkflow':'Missing_IDs From Workflow (Plasmid)'})
banchmarkingFinal
```

```python
banchmarkingFinal = banchmarkingFinal.replace({'','NaN,'}, np.nan) # some of the empty spaces were not refelecting NAN so retrying the command to achive the goal
banchmarkingFinal
```
